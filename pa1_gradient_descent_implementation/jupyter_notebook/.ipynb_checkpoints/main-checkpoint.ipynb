```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#constants
file_path = "../data.txt"
```

```python
def read_data(file_path):
    #open file and read data
    file = open(file_path, 'r')
    data = file.read()
    split_xy_data = data.split("\n")
    #removing the last endline
    split_xy_data.pop()
    print(split_xy_data)
    #split data into separate x and y array
    x = [xy.split(',')[0] for xy in split_xy_data]
    y = [xy.split(',')[1] for xy in split_xy_data]
    
    return x, y

def plot_data(x, y):
    plt.plot(x, y)
    plt.show()
```

```python
x_values, y_values = read_data(file_path)
```
